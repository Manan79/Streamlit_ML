```python
import pandas as pd
```

```python
data = pd.read_csv('Housing.csv')
```

```python
data.drop(['hotwaterheating' , 'airconditioning' , 'prefarea'] , axis = 1 , inplace= True)
```

```python
data
```

```python
new_data = data.loc[:,['area' , 'bedrooms'  , 'price']]
new_data.dropna()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler= StandardScaler()
```

```python
scaler.fit(new_data.loc[:,['area' , 'bedrooms'  ]])
```

```python
scaler_data = scaler.transform((new_data.loc[:,['area' , 'bedrooms' ]]))
```

```python
new_data.loc[:,['area' , 'bedrooms' ]] = scaler_data
```

```python
new_data
```

```python
x=new_data.iloc[:,:2]
y=new_data.iloc[:,2]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train, y_test= train_test_split(x,y,test_size=0.25)
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model= LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model.predict(x_test)
```

```python
y_pred= model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
print("Mean Squared Error is",mean_squared_error(y_test,y_pred))
print("Mean Absolute Error is",mean_absolute_error(y_test,y_pred))
print("R2 Score is",r2_score(y_test,y_pred))
```
